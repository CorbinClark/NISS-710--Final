### Introduction

Natural disasters are unavoidable exogenous shocks that have a profound impact on how we live, and the effects can be felt long after reconstruction has finished.  A city built on a fault line like San Francisco, founded in a dry climate promoting wildfires, or in a tropical environment on the coast generating severe storms and hurricanes, is shaped by the elements.  The institutions that are present in a region (in the context of this paper, state) are morphed by natural disasters as well.  In transaction cost economics, an exogenous shock impeding the mechanisms that facilitate transactions would constitute institutional change (Kingston and Caballero, 2009).  Should the existing institutions fail to be efficient, new institutions better adapted to the landscape will emerge.  To study this further, this paper focuses on to what extent property damage caused by a natural disaster impacts economic freedom, measured by Fraser.  Support for the selection of this index will be provided in the theory section of this paper.  In addition to the Fraser economic freedom score, the number of construction workers will be observed within the same framework.  

   Discussion in the New Institutional field has explored how to measure formal institutions to a great extent, but the answer to measuring informal institutions has been elusive. It is critical to understand both informal and formal institutions to properly analyze a society. This paper looks at past methods of analysis regarding both the measuring of formal and informal institutions, keeping in mind natural disasters to determine if they have an impact on the strength of institutions prior to, or after a catastrophe. By using a panel dataset of all 50 states, as well as the District of Columbia, dating back to 1997, patterns emerge, leading to conclusions regarding institutions in place in states that have a high rate of natural disasters as well as those that do not experience natural disasters frequently. Property damage is used in this paper as a measure of disaster, adjusted for inflation and logged due to the initial skewness of the residuals.  Fixed-effects models are used to control for time-invariant variables; several outcome variables are observed to identify changes in formal and informal institutions due to natural disasters (property damage). This paper will examine prior research on the impact that external shocks have on institutions and relate that back to how natural disasters fit in that framework, with support coming from the fixed-effects regression.  The empirical analysis will focus on the entirety of the United States, drawing from the aforementioned panel data and extensive disaster data granted by Arizona State University.

   This paper also incorporates transaction cost theory to examine how a natural disaster decreases institutional strength. As evidenced in previous literature that will be covered in the literature review section of this paper, transaction costs and institutions are linked and have a close relationship that shows that institutions are created to reduce transaction costs.  They lower uncertainty and risk by limiting individual freedom of action, and natural disasters disrupt institutions and increase these transaction costs.   

This paper is a preliminary paper designed to bolster a thesis paper regarding economic growth and institutions that utilizes the synthetic control method to determine the impact that natural disasters have on the economy. ### Literature Review

In Douglass North’s seminal work (1984), the link between institutions and transaction costs is explored, and will be examined by researchers for the decades to follow. His work, which was created to fill a perceived hole in the existing literature regarding institutions, was dedicated to showing that in real-world situations, there are costs that must be incurred when allocating resources; transaction costs. Transaction costs are present in every exchange, and as North argues, discussing economic theory without the inclusion of such costs is fruitless. As Raschky (2008) finds, countries with strong institutions experience less victims and significantly lower economic loss from natural disasters than countries with weak institutions.  The function of this paper is focused on the flip side of this coin, which is notably missing from the current new institutional literature, and therefore a theory of the relationship between the two must be formulated.   

Natural disasters are unpredictable and unstoppable, and thus, they play a significant role in any region that experiences them. Hurricanes are especially impactful, as they occur relatively frequently, with 2020 alone seeing six major storms making landfall on the United States (Masters, 2020). There are studies that show that the economic impact of a disaster is negligible or even has a positive effect on the economy (Park and Wang, 2017; Barone and Mocetti, 2014), but there are several problems with their methods that may contribute to their findings. A disaster, be it natural or man-made, provides social scientists an excellent opportunity to observe society at its most exposed. Because disasters lay bare the workings of both formal and informal institutions, they provide us with opportunities for a closer examination of the institutions at work (Rayamajhee, 2020). This allows us to investigate to what extent natural disasters are related to formal and informal institutions. 

Acemoglu and Johnson (2005) present property rights-protecting institutions as those aimed at constraining the government from engaging in public predation, and stress their importance. On the other hand, Williamson and Kerekes (2011) find that informal property securing institutions matter more than the formal ones (also see Benson, 1989). In Acemoglu et al. (2001), institutions and their concepts are examined as they relate to colonization policies undertaken by European countries. In their work, they discover that extractive institutions lead to “vicious circles,” wherein the institutions are created by the people in power to benefit those very same people. These “vicious circles” are explained in great detail in Acemoglu and Robinson’s book Why Nations Fail along with their counterpart “virtuous circles”. This work by Acemoglu, Johnson, and Robinson exposes reasons for certain country’s economic growth in relation to others based on institutions. This relates to the subject of natural disasters in an important manner.  The society that is left bare in the wake of a disaster mimics this environment and is a breeding ground for institutional change (Yamamura, 2014). Governing bodies are often in search of manners in which to mitigate disasters and stabilize the region, oftentimes landing on policy change that is intended to secure property-rights and reduce transaction costs, and thus strengthen institutions (Liang and Cao, 2015; Smith, Sandler, and Goralnik, 2013).  Through this context, a natural disaster would not have a strong effect on institutions, because the institutions in place are designed to weather both the literal and metaphorical storm.   

For this paper, it is important to also understand how social capital and enforcement costs, which are included in transaction costs, interact, specifically in a society after a disaster.  Social capital and enforcement costs have an inverse relationship where an increase (decrease) in social capital leads to a decrease (increase) in enforcement costs.  Grafton (2005) demonstrates this relationship using social capital within the governance of fisheries. Grafton finds that when there is higher social capital among the fishers; measured in trust, civic engagement and co-operation, and social networks, then there is a reduced ‘race to fish’ and requires less governance. The article also finds that monitoring and enforcement costs for regulators may be reduced should fishers increase their compliance, therefore raising social capital. Landry et al. (2002) claims that the dominating view of social capital, that it cannot be captured through a single indicator, and uses various measures and explanatory variables to gauge the level of social capital. The five forms of structural social capital they choose are: business network assets, information network assets, research network assets, participation assets, and relational assets, along with a single form of cognitive social capital: reciprocal trust. The conclusion of this article states that social capital contributes to innovation by reducing transaction costs between firms and other actors, notably policing and enforcement costs. This conclusion is supported through Maskell (2000), where social capital and said transaction costs are linked further. 

In this paper, social capital is examined through murder rate, a variable included in the panel data gathered for this paper.  Murder rate is used as an inverse proxy for social capital; an increase (decrease) in murder rate would be a decrease (increase) in social capital.  This variable is used because there was a strong inverse relationship and correlation between murder rate and social capital in a study conducted by Putnam (1996, 2000).  Putnam comes at this issue from the reverse angle, looking at how social capital predicts murder rate, and he finds that it is the "strongest predictor", higher than poverty rate and other plausible measures.  

### Theory

Institutional strength is built up over many years of the previously mentioned “virtuous circle” and competent policies along with accountability. The tearing down of sturdy institutions has a wider variety of methods and can be much swifter. A natural disaster theoretically decreases the strength of both formal and informal institutions in several ways. Formal institutions can be weakened by corrupt individuals that oversee said institutions. Natural disasters provide fertile ground for extractive institutions to take root and further, activities such as corruption and property theft are made more attractive due to the lack of reinforcement (Testa, 2021). In the wake of a natural disaster such as a hurricane, there are more pressing matters at hand than holding individuals accountable for poor policies and what seem like minor transgressions. These small transgressions can oftentimes snowball into a much larger problem, the “vicious circle.”  When strong institutions are in place prior to a disaster, the small malfeasances are weeded out and caught earlier, stopping the ball in its tracks. Therefore, having a robust system of checks and balances can prevent the diminishment of formal institutions. 

The output variable of Fraser Economic Freedom is chosen because it shows the strength of formal political institutions.  The focus of the majority of scholars regarding government spending and institutions is the effectiveness of the budget and not the size (Garayeva, Aygun and Tahirova, Gulzar, 2016).  The Fraser index takes into account a multitude of factors, including government expenditures, taxation, labor market restrictions along with other economic indicators. The index is measured on a scale of one to ten, with no state having higher than 8.5 or lower than 4.  

Informal institutions can similarly deteriorate in the wake of a natural disaster, but through different avenues. Informal institutions are unwritten rules of a society that are constructed, transmitted, and upheld through social interactions and are enforced outside the typical channels. Due to the unstructured nature of these institutions, a strong disruption in the community could knock them off balance. A natural disaster acts as just such a disruption for social capital. Communities that are displaced by a natural disaster lose the ability to contact the other community members and therefore lack the methods of monitoring the actions of one another and upholding the socially shared rules that governed their interactions, and therefore theoretically lose social capital. In especially severe cases, natural disasters scatter a community, and destroy public (and private) modes of transportation, further limiting the capacity for individuals to interact with other members of their community. Once the disaster has come and passed, the work needed to rebuild the area may create stronger social capital, and new informal institutions. Unfortunately, there is yet to be a reliable index for social capital, so in this paper, social capital is measured through the murder rate of a state.  According to Putnam(1996, 2000), social capital is the strongest indicator for murder rate, and they are highly correlated, as well as being inversely linearly related. In this paper, the murder rate will be used as the outcome variable for the fixed-effects model that is meant to represent social capital, or informal institutions.

Typically, the approach to overcome heterogeneity issues is to use panel data and perform either a random or fixed-effect estimation.  Because individual heterogeneity is viewed as stationary across time, the observation of individual states at different time points allows researchers to control for the heterogeneous factors without having to directly observe and/or quantify them. When important explanatory variables are left out of an OLS, because of lack of data, or other reasons rendering them unavailable, there will be biased estimates.  Ferrer-I-Carbonell and Frijters (2004) show that this bias for not controlling for heterogeneity on an individual level and the positive coefficient reduces by roughly 2/3 when going from a pooled OLS to a fixed-effect estimator. They further demonstrate the importance of controlling for important yet unobservable characteristics. Using fixed effect estimation, there is a narrowing of analysis to specific dimensions of the data available. The tradeoff here is with bias-variance; one forgoes the opportunity to explain between-unit differences in outcomes in exchange for a bias-free estimate of a within-unit treatment effect.  Identifying which units show significant variation over time is important to note as well for gauging the generalizability of the results.  The formal standard fixed effects dummy variable model is as follows: 
#### Yit = β0 + αi + β1Xit + ε it
where the outcome Y and an independent variable(s) or treatment X are observed for every unit (state) i over multiple time periods (years) t, and mutually exclusive intercept shift, α, are estimated for each unit i to capture the distinctive, time-invariant features of each unit. This results in an estimate of β1 that is purged of the influence of between-unit time-invariant confounders. 

```python
import statsmodels.api as sm
from linearmodels import PanelOLS

def panel_regression(reg_df, x, y, num_exog = 1):
#results = model.fit(cov_type='HC1')
    if num_exog == 1:
        X = reg_df[[x]]
        X["Constant"] = 1
        Y = reg_df[[y]]
        model = PanelOLS(Y, X, entity_effects=True, time_effects=False)
        results = model.fit(cov_type='clustered', cluster_entity=True)
        reg_df["predict:" + y + "~" + x] = results.predict()
#reg_df["residual:" + y + "~" + x] = reg_df[y] - results.predict()
        
        print(results)
        
        fig, ax = plt.subplots(figsize = (20,14))
        reg_df.plot.scatter(x=x, y=y, ax=ax)
        reg_df.plot.line(x=x, y = "predict:" + y + "~" + x, ax=ax)
    else:
        X = reg_df[x]
        X["Constant"] = 1
        Y = reg_df[y]
        model = PanelOLS(Y,X, entity_effects=True, time_effects=False)
        results = model.fit(cov_type = 'clustered', cluster_entity=True)
        reg_df["predict:" +y[0]] = results.predict()
#reg_df["residual:" + y + "~" + x] = reg_df[y] - results.predict()

        print(results)
    
        fig, ax = plt.subplots(figsize = (20,14))
        reg_df.plot.scatter(x=y[0], y = "predict:" + y[0], ax = ax)
        
    plt.show()
    plt.close()
```

### Conclusion

Measuring institutions, both formal and informal, is difficult for a variety of reasons, including the nebulous nature of the “rules of a game.”  To narrow this topic down, this paper has selected certain aspects of institutions from the previous research and used those to analyze the impact that a natural disaster has on institutions. The models used fail to show the stated hypothesis that as the severity of natural disasters increase, so does the strength of formal and informal institutions.

Both the fixed-effects regression of Fraser index and murder rate were deemed inconclusive.  The coefficients followed the logic presented in the theory and literature review sections of the paper, but the p-values were too high to draw any significant conclusions.  This is in line with the literature that states that countries (in this case it is states) with strong institutions are designed to withstand shocks, both exogenous and endogenous (Yamamura, 2014).

As noted before, there is no perfect variable that models institutions, be they formal or informal.  Informal institutions specifically are difficult to measure, with Putnam (1996, 2000) admitting that there is no combined metric that is great. Hence, the outcome variables in the models, Fraser Economic Freedom and murder rate respectively, are not flawless proxies for this study, although I believe that the arguments made in this paper do justify their use in this context.  As mentioned previously, natural disasters impact countries with weak institutions more than countries with strong institutions, and all of the United States have relatively strong institutions, and are less likely to suffer from institutional change.  This study could be undertaken on a broader scale for countries with weaker institutions and the results may be found to be statistically significant.  Another possible limitation to this study could be that there are important predictor variables that were not accounted for in the regressions.  This could lead to the regression failing to capture the entirety of what is occurring when a disaster strikes.  That being said, I believe this study supports what little literature exists in the field regarding natural disasters and institutions, and may be expanded upon in the future to observe other countries at the state/province/region-level, or on a larger, international scale.
